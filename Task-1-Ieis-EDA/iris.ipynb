# Iris Dataset

# 1. Introduction

The objective of this task is to explore and visualize the famous Iris dataset in order to understand its structure, distributions, and relationships between features. This task focuses on building foundational data science skills such as data loading, inspection, and visualization using Python.

# 2. Dataset Description

The Iris dataset contains measurements of iris flowers from three different species:

Setosa

Versicolor

Virginica

Features:

Sepal Length

Sepal Width

Petal Length

Petal Width

Target Variable:

Species

# 3. Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 4. Load the Dataset

```python
iris = sns.load_dataset('iris')
```

# 5. Dataset Inspection

```python
iris.head()
iris.shape
iris.columns
```

Insight:

The dataset contains 150 rows and 5 columns.

Four columns are numerical features, and one column represents the flower species.

The dataset is clean and well-structured.

# 6. Summary Statistics

```python
iris.describe()
```

# Insight:
Petal length and petal width show higher variation compared to sepal features.
This suggests petal features may be more useful for distinguishing species.

# 7. Data Visualization (EDA)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width')
plt.show()
```

# Insight:

Clear separation between species is visible.
Setosa is distinctly separated, while Versicolor and Virginica show some overlap

```python
plt.figure(figsize=(6,4))
plt.hist(iris['sepal_length'], bins=20)
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

# Insight:

Sepal length follows an approximately normal distribution.
Most values fall between 5.0 and 6.5.

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Petal Length by Species')
plt.show()
```

# Insight:

Setosa has the smallest petal length with minimal variation.
Virginica shows the highest median petal length.
Few outliers are present, indicating stable measurements

# 8. Conclusion

The Iris dataset is clean and well-balanced.

Petal features are more effective than sepal features in distinguishing species.

Visualization techniques like scatter plots and box plots provide valuable insights into feature relationships.
